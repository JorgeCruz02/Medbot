````python
#Creamos un nuevo data frame donde agregaremos los datos del conteo de P1 y P2
Comparacion_conteo_elementos_df = conteo_elementos_P1_df.copy()

#Agregamos la columna de frecuencia al nueva DataFrame
Comparacion_conteo_elementos_df['Frecuencia_P2'] = conteo_elementos_P2_df['Frecuencia']

#Renomnbramos las columnas del nuevo data frame
nuevos_nombres = {'Valor': 'Mes', 'Frecuencia': 'Frecuencia_P1'}
Comparacion_conteo_elementos_df.rename(columns = nuevos_nombres, inplace=True)
#Mostramos el Nuevo DataFrame
Comparacion_conteo_elementos_df
````

Ordenamos el data frame.

````python
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
Dias = [31,28, 31,30, 31,30, 31,31, 30,31, 30,31]
#Creamos un diccionario de datos en el cual agregaremos el mes junto con la suma total de los pasos echos ese mes
Comparacion_conteo_elementos = {'Mes':[], 'Frecuencia_P1':[], 'Porcentaje_P1':[],'Frecuencia_P2':[],'Porcentaje_P2':[]}
for mes in meses:
  P1=conteo_elementos_P1_df[conteo_elementos_P1_df['Valor'] == mes]
  P2=conteo_elementos_P2_df[conteo_elementos_P2_df['Valor'] == mes]
  Comparacion_conteo_elementos['Mes'].append(mes)
  Comparacion_conteo_elementos['Frecuencia_P1'].append(P1['Frecuencia'].sum())
  Comparacion_conteo_elementos['Frecuencia_P2'].append(P2['Frecuencia'].sum())

a = 0
for dia in Dias:
    Porcentaje1 = (Comparacion_conteo_elementos['Frecuencia_P1'][a]*100)/dia
    Porcentaje2 = (Comparacion_conteo_elementos['Frecuencia_P2'][a]*100)/dia
    Comparacion_conteo_elementos['Porcentaje_P1'].append(round(Porcentaje1, 2))
    Comparacion_conteo_elementos['Porcentaje_P2'].append(round(Porcentaje2, 2))
    a = a + 1

Comparacion_conteo_elementos = pd.DataFrame(Comparacion_conteo_elementos)
Comparacion_conteo_elementos

````
