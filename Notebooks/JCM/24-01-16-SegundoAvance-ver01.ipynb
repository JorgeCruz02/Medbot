````python
from datetime import datetime, timedelta
import numpy as np

# Convercion de la columan 'startTime' a datetime.
DataPerson01_A_M['time'] = pd.to_datetime(DataPerson01_A_M['time'])

#creamos un ciclo donde recorreremos el tata frame 'DataPerson01_A_M'
for index, row in DataPerson01_A_M.iterrows():
  #Para el primer renglon craremos un data frame auxiliar donde tendremos 2 columnas nuevas 'start_time'
  #y 'end_tiem' dejando del lado la columna 'time'.
  if(index==0):
    data2 = {'date': [row['date']],
            'start_tiem': [row['time']],
            'end_time': [row['time']],
            'steps': [row['steps']]}
    df2 = pd.DataFrame(data2)
  else:
    Tiempo_anterior =  df2.iloc[-1]['end_time']
    # Agregar 1 minuto o 60 segundos
    Tiempo_anterior = Tiempo_anterior + np.timedelta64(60, 's')
    Nuevo_tiempo = row['time']
    if(Nuevo_tiempo == Tiempo_anterior):
      df2.iloc[-1]['end_time'] = row['time']
      df2.iloc[-1]['steps'] = df2.iloc[-1]['steps'] + row['steps']
    else:
      Nuevo_renglon = {'date': [row['date']],
                      'start_tiem': [row['time']],
                      'end_time': [row['time']],
                      'steps': [row['steps']]}
      df2.append(Nuevo_renglon, ignore_index=True)

df2.head(5)
````
